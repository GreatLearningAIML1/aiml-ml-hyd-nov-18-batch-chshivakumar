### Find out the no of male & female employees in the company

*Hint: Use can use a condition to find out the no of 1s and 0s in gender and sum it*

```python

print("No of Males ", hr[hr['Gender']==1].shape[0])
```

```python
print("No of Females ", hr.shape[0]-hr[hr['Gender']==1].shape[0])
```

### Find out the IQR for the column - Age 

*Hint: IQR - Interquartile range = 75th percentile - 25th percentile* 

```python
print("IQR of Age ",hr['Age'].quantile(0.75) -hr['Age'].quantile(0.25))
```

### Find out the mean, median and variance for the column - DailyRate 

```python
print("Mean of DailyRate ", hr['DailyRate'].mean())
```

```python
print("Median of DailyRate", hr['DailyRate'].median())
```

```python
print("Variance of DailyRate", hr['DailyRate'].var())
```
